```python
import sqlite3
conn = sqlite3.connect('nhahang.db')
cursor = conn.cursor()

# To bng CUAHANG
cursor.execute('''
CREATE TABLE IF NOT EXISTS CUAHANG (
    MaCH INT PRIMARY KEY,
    TenCH VARCHAR(20) NOT NULL,
    GioMoCua TIME NOT NULL,
    GioDongCua TIME NOT NULL
)
''')

# To bng LOAIMON
cursor.execute('''
CREATE TABLE IF NOT EXISTS LOAIMON (
    MaLM INT PRIMARY KEY,
    TenLM VARCHAR(20) NOT NULL
)
''')

# To bng MONAN
cursor.execute('''
CREATE TABLE IF NOT EXISTS MONAN (
    MaMA INT PRIMARY KEY,
    TenMA INT NOT NULL,
    MaLM INT NOT NULL,
    GiaBan INT NOT NULL,
    FOREIGN KEY (MaLM) REFERENCES LOAIMON(MaLM)
)
''')

# To bng KHACHHANG
cursor.execute('''
CREATE TABLE IF NOT EXISTS KHACHHANG (
    MaKH INT PRIMARY KEY,
    HoTen VARCHAR(20) NOT NULL
)
''')

# To bng DATMON
cursor.execute('''
CREATE TABLE IF NOT EXISTS DATMON (
    MaDM INT PRIMARY KEY,
    MaMA INT NOT NULL,
    SoLuong INT NOT NULL,
    FOREIGN KEY (MaMA) REFERENCES MONAN(MaMA)
)
''')

# To bng DANHGIA
cursor.execute('''
CREATE TABLE IF NOT EXISTS DANHGIA (
    MaDG INT PRIMARY KEY,
    MaKH INT NOT NULL,
    MaCH INT NOT NULL,
    SoSao INT NOT NULL CHECK (SoSao >= 0 AND SoSao <= 5),
    NgayDanhGia DATE NOT NULL,
    FOREIGN KEY (MaKH) REFERENCES KHACHHANG(MaKH),
    FOREIGN KEY (MaCH) REFERENCES CUAHANG(MaCH)
)
''')
```

```python
query = '''
SELECT C.TenCH, L.TenLM, SUM(M.GiaBan * D.SoLuong) AS DoanhThu
FROM CUAHANG C
JOIN DANHGIA DG ON C.MaCH = DG.MaCH
JOIN MONAN M ON 1=1  
JOIN LOAIMON L ON M.MaLM = L.MaLM
JOIN DATMON D ON M.MaMA = D.MaMA
GROUP BY C.MaCH, L.MaLM
ORDER BY C.MaCH, L.MaLM
'''
for row in cursor.execute(query):
    print(row)
```

```python
query = '''
SELECT C.TenCH, AVG(DG.SoSao) AS TB_Sao,
    CASE
        WHEN AVG(DG.SoSao) < 3.5 THEN 'Can cai thien'
        WHEN AVG(DG.SoSao) BETWEEN 3.5 AND 4.5 THEN 'Tot'
        ELSE 'Xuat sac'
    END AS Muc_Hai_Long
FROM CUAHANG C
JOIN DANHGIA DG ON C.MaCH = DG.MaCH
GROUP BY C.MaCH
'''
for row in cursor.execute(query):
    print(row)
```
