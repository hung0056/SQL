```python
import sqlite3
conn = sqlite3.connect('nhahang.db')
cursor = conn.cursor()

# Tạo bảng CUAHANG
cursor.execute('''
CREATE TABLE IF NOT EXISTS CUAHANG (
    MaCH INT PRIMARY KEY,
    TenCH VARCHAR(20) NOT NULL,
    GioMoCua TIME NOT NULL,
    GioDongCua TIME NOT NULL
)
''')

# Tạo bảng LOAIMON
cursor.execute('''
CREATE TABLE IF NOT EXISTS LOAIMON (
    MaLM INT PRIMARY KEY,
    TenLM VARCHAR(20) NOT NULL
)
''')

# Tạo bảng MONAN
cursor.execute('''
CREATE TABLE IF NOT EXISTS MONAN (
    MaMA INT PRIMARY KEY,
    TenMA INT NOT NULL,
    MaLM INT NOT NULL,
    GiaBan INT NOT NULL,
    FOREIGN KEY (MaLM) REFERENCES LOAIMON(MaLM)
)
''')

# Tạo bảng KHACHHANG
cursor.execute('''
CREATE TABLE IF NOT EXISTS KHACHHANG (
    MaKH INT PRIMARY KEY,
    HoTen VARCHAR(20) NOT NULL
)
''')

# Tạo bảng DATMON
cursor.execute('''
CREATE TABLE IF NOT EXISTS DATMON (
    MaDM INT PRIMARY KEY,
    MaMA INT NOT NULL,
    SoLuong INT NOT NULL,
    FOREIGN KEY (MaMA) REFERENCES MONAN(MaMA)
)
''')

# Tạo bảng DANHGIA
cursor.execute('''
CREATE TABLE IF NOT EXISTS DANHGIA (
    MaDG INT PRIMARY KEY,
    MaKH INT NOT NULL,
    MaCH INT NOT NULL,
    SoSao INT NOT NULL CHECK (SoSao >= 0 AND SoSao <= 5),
    NgayDanhGia DATE NOT NULL,
    FOREIGN KEY (MaKH) REFERENCES KHACHHANG(MaKH),
    FOREIGN KEY (MaCH) REFERENCES CUAHANG(MaCH)
)
''')
```

```python
query = '''
SELECT C.TenCH, L.TenLM, SUM(M.GiaBan * D.SoLuong) AS DoanhThu
FROM CUAHANG C
JOIN DANHGIA DG ON C.MaCH = DG.MaCH
JOIN MONAN M ON 1=1  
JOIN LOAIMON L ON M.MaLM = L.MaLM
JOIN DATMON D ON M.MaMA = D.MaMA
GROUP BY C.MaCH, L.MaLM
ORDER BY C.MaCH, L.MaLM
'''
for row in cursor.execute(query):
    print(row)
```

```python
query = '''
SELECT C.TenCH, AVG(DG.SoSao) AS TB_Sao,
    CASE
        WHEN AVG(DG.SoSao) < 3.5 THEN 'Can cai thien'
        WHEN AVG(DG.SoSao) BETWEEN 3.5 AND 4.5 THEN 'Tot'
        ELSE 'Xuat sac'
    END AS Muc_Hai_Long
FROM CUAHANG C
JOIN DANHGIA DG ON C.MaCH = DG.MaCH
GROUP BY C.MaCH
'''
for row in cursor.execute(query):
    print(row)
```

```python
#Tính ngưỡng ngoại lai
query = '''
WITH Q AS (
  SELECT
    (SELECT SoLuong FROM DATMON ORDER BY SoLuong LIMIT 1 OFFSET (SELECT COUNT(*) * 1 / 4 FROM DATMON)) AS Q1,
    (SELECT SoLuong FROM DATMON ORDER BY SoLuong LIMIT 1 OFFSET (SELECT COUNT(*) * 3 / 4 FROM DATMON)) AS Q3
),
Bounds AS (
  SELECT
    Q1 - 1.5 * (Q3 - Q1) AS lower_bound,
    Q3 + 1.5 * (Q3 - Q1) AS upper_bound
  FROM Q
)
SELECT lower_bound, upper_bound FROM Bounds;
'''
cursor.execute(query)
lower_bound, upper_bound = cursor.fetchone()
print(f'Ngưỡng ngoại lai: [{lower_bound}, {upper_bound}]')

# Lấy bản ghi ngoại lai
query = '''
WITH Q AS (
  SELECT
    (SELECT SoLuong FROM DATMON ORDER BY SoLuong LIMIT 1 OFFSET (SELECT COUNT(*) * 1 / 4 FROM DATMON)) AS Q1,
    (SELECT SoLuong FROM DATMON ORDER BY SoLuong LIMIT 1 OFFSET (SELECT COUNT(*) * 3 / 4 FROM DATMON)) AS Q3
),
Bounds AS (
  SELECT
    Q1 - 1.5 * (Q3 - Q1) AS lower_bound,
    Q3 + 1.5 * (Q3 - Q1) AS upper_bound
  FROM Q
)
SELECT *
FROM DATMON, Bounds
WHERE SoLuong < lower_bound OR SoLuong > upper_bound;
'''
cursor.execute(query)
outliers = cursor.fetchall()
print('Bản ghi ngoại lai:', outliers)

# Xóa bản ghi ngoại lai
query = '''
WITH Q AS (
  SELECT
    (SELECT SoLuong FROM DATMON ORDER BY SoLuong LIMIT 1 OFFSET (SELECT COUNT(*) * 1 / 4 FROM DATMON)) AS Q1,
    (SELECT SoLuong FROM DATMON ORDER BY SoLuong LIMIT 1 OFFSET (SELECT COUNT(*) * 3 / 4 FROM DATMON)) AS Q3
),
Bounds AS (
  SELECT
    Q1 - 1.5 * (Q3 - Q1) AS lower_bound,
    Q3 + 1.5 * (Q3 - Q1) AS upper_bound
  FROM Q
)
DELETE FROM DATMON
WHERE SoLuong < (SELECT lower_bound FROM Bounds)
   OR SoLuong > (SELECT upper_bound FROM Bounds);
'''
cursor.execute(query)
conn.commit()
print('Đã xóa bản ghi ngoại lai')
conn.close()
```
